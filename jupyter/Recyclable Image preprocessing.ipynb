```python
#image rgb to grayscale, CLAHE n back to 3 channel
import cv2
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# j = get num of image number in folder
j = len(os.listdir('D:\\Users\\Adib Razak\\Desktop\\Recycleable_Classification\\dataSetProcessed\\smooth\\NonRecyclable'))

for i in range(j):
    # Reading image from directory
    img = cv2.imread('D:/Users/Adib Razak/Desktop/Recycleable_Classification/dataSetProcessed/smooth/NonRecyclable/avgN'+str(i)+'.jpg')
    averaging = cv2.blur(img, (10, 10))
 
    # Resizing image for compatibility
    img = cv2.resize(img, (500, 600))

    # The initial processing of the image
    # img = cv2.medianBlur(image, 3)
    image_bw = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # The declaration of CLAHE
    # clipLimit -> Threshold for contrast limiting
    clahe = cv2.createCLAHE(clipLimit = 5)
    clahe_img = clahe.apply(image_bw) + 30

    #Convert to 3 channel
    final_img = cv2.cvtColor(clahe_img, cv2.COLOR_GRAY2RGB)
    
    cv2.imwrite('D:/Users/Adib Razak/Desktop/Recycleable_Classification/dataSetProcessed/smoothClahe/NonRecyclable/avgClhN'+str(i)+'.jpg',final_img)
```

```python
#live camera edge detection
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 30, 100)
    cv2.imshow("edges", edges)
    cv2.imshow("gray", gray)
    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#canny edge detection
#The smallest value between threshold1 and threshold2 is used for edge linking. 
#The largest value is used to find initial segments of strong edges

import cv2
import numpy as np
import matplotlib.pyplot as plt

# read the image
image = cv2.imread('D:/Users/Adib Razak/Desktop/Academic/Sem 10/Imaging System (ECC4403)/Assignments Group Project/overUnderExposed/bottleUnderExposed.jpg')

# convert it to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#show grayscale image
#plt.imshow(gray, cmap="gray")
#plt.show()

# perform the canny edge detector to detect image edges
edges = cv2.Canny(gray, threshold1=10, threshold2=100)

# show the image edges
plt.imshow(edges, cmap="gray")
plt.show()
```

```python
#image brightness + smooth + canny edge detection
#The smallest value between threshold1 and threshold2 is used for edge linking. 
#The largest value is used to find initial segments of strong edges

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance

# read the image
img = Image.open('D:/Users/Adib Razak/Desktop/Academic/Sem 10/Imaging System (ECC4403)/Assignments Group Project/overUnderExposed/bottleUnderExposed.jpg')

#enhance brightness
img_con = ImageEnhance.Brightness(img)
bright_img = img_con.enhance(3)

###convert PIL img to OpenCV img will able to keep transparent channel
#convert PIL img to OpenCV img
pil_image = bright_img.convert('RGB') 
open_cv_image = np.array(pil_image) 
####

#smooth image
gaussian = cv2.GaussianBlur(open_cv_image, (21, 21), 0)

# Convert RGB to BGR 
gaussian = open_cv_image[:, :, ::-1].copy() 

#convert it to grayscale
gray = cv2.cvtColor(gaussian, cv2.COLOR_BGR2GRAY)

#show grayscale image
#plt.imshow(gray, cmap="gray")
#plt.show()

# perform the canny edge detector to detect image edges
edges = cv2.Canny(gray, threshold1=30, threshold2=150)

# show the image edges
plt.imshow(edges, cmap="gray")
plt.show()
```

```python
#image brightness + smooth + canny edge detection
#The smallest value between threshold1 and threshold2 is used for edge linking. 
#The largest value is used to find initial segments of strong edges

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance


# j = get num of image number in folder
j = len(os.listdir('D:\\Users\\Adib Razak\\Desktop\\all'))

# read the image
for i in range(j):
    img = Image.open('D:/Users/Adib Razak/Desktop/all/img_'+str(i+1)+'.jpg')
        
    # read the image
    #img = Image.open('D:/Users/Adib Razak/Desktop/Academic/Sem 10/Imaging System (ECC4403)/Assignments Group Project/overUnderExposed/bottleUnderExposed.jpg')
    
    #enhance brightness
    img_con = ImageEnhance.Brightness(img)
    bright_img = img_con.enhance(3)
    
    ###convert PIL img to OpenCV img will able to keep transparent channel
    #convert PIL img to OpenCV img
    pil_image = bright_img.convert('RGB') 
    open_cv_image = np.array(pil_image) 
    ####
    
    #smooth image
    gaussian = cv2.GaussianBlur(open_cv_image, (21, 21), 0)
    
    # Convert RGB to BGR 
    bgr = open_cv_image[:, :, ::-1].copy() 
    
    #convert it to grayscale
    gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
    
    #show grayscale image
    #plt.imshow(gray, cmap="gray")
    #plt.show()
    
    # perform the canny edge detector to detect image edges
    edges = cv2.Canny(gray, threshold1=30, threshold2=200)
    
    # show the image edges
    #plt.imshow(edges, cmap="gray")
    #plt.show()
    
    #Convert to 3 channel
    final_img = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)
    
    cv2.imwrite('D:/Users/Adib Razak/Desktop/all_processed/img_'+str(i+1)+'.jpg',final_img)
```
